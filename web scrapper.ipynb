```python
import newspaper
```

```python
paper = newspaper.build('https://nypost.com/')
```

```python
print(paper.size())
```

```python
for article in paper.articles:
  print(article.url)
```

```python
for category in paper.category_urls():
  print(category)
```

```python
print("brand:", paper.brand)
print("Description:",paper.description)
```
